From problem 1, I got  

$\begin{cases}
\dot{\chi}=\frac{\dot{\phi}_{0}^{2}\sin^{4}\theta_{0}\cos\theta}{\sin^{3}\theta}\\
\dot{\theta}=\chi\\
\dot{\phi}=\frac{\dot{\phi}_{0}\sin^{2}\theta_{0}}{\sin^{2}\theta}
\end{cases}$

Now, let $u=\cos\left[\theta\left(t\right)\right]$

$\dot{u}=-\dot{\theta}\sin\left[\theta\left(t\right)\right]$

$\ddot{u}=-\ddot{\theta}\sin\left[\theta\left(t\right)\right]-\left(\dot{\theta}\right)^{2}\cos\left[\theta\left(t\right)\right]$

$-\ddot{\theta}\sin\left[\theta\left(t\right)\right]=-\frac{\dot{\phi}_{0}^{2}\sin^{4}\theta_{0}u}{1-u^{2}}$

$\ddot{u}=-\frac{\dot{\phi}_{0}^{2}\sin^{4}\theta_{0}u}{1-u^{2}}-\left(\dot{\theta}\right)^{2}u$

because $\left(\dot{\theta}\right)^{2}=\left(\dot{\theta}_{0}\right)^{2}+\dot{\phi}_{0}^{2}\left(\sin^{2}\theta_{0}-\frac{\sin^{4}\theta_{0}}{1-u^{2}}\right)$

$\ddot{u}=-\frac{\dot{\phi}_{0}^{2}\sin^{4}\theta_{0}u}{1-u^{2}}-\left\{ \left(\dot{\theta}_{0}\right)^{2}+\dot{\phi}_{0}^{2}\left(\sin^{2}\theta_{0}-\frac{\sin^{4}\theta_{0}}{1-u^{2}}\right)\right\} u$

$\ddot{u}=-\left(\dot{\theta}_{0}^{2}+\dot{\phi}_{0}^{2}\sin^{2}\theta_{0}\right)u$

$u=A\cos\left[\sqrt{\dot{\theta}_{0}^{2}+\dot{\phi}_{0}^{2}\sin^{2}\theta_{0}}t\right]$

Because $\dot{\theta}_{0}^{2}+\dot{\phi}_{0}^{2}\sin^{2}\theta_{0}=\frac{V_{0}^{2}}{R^{2}}$, it can also be written as

$u=A\cos\left[\frac{V_{0}}{R}t\right]$

Therefore, the period is $T=\frac{2\pi R}{V_{0}}$. 




```python
import numpy as np
def T( dphi0, dthet0, theta0):
    return 2*np.pi/np.sqrt((dphi0*np.sin(theta0))**2+dthet0**2)

print('The period of the first case in problem 1 is',T( np.pi/10, np.pi/10, np.pi/18))
print('The period of the second case in problem 1 is',T( np.pi/10, np.pi/10, np.pi/3))
print('The period of the third case in problem 1 is',T( np.pi/5, np.pi/5, np.pi/3))
print('The period of the fourth case in problem 1 is',T( np.pi/10, np.pi/5, np.pi/3))
```

The case under the gravity
\begin{cases}
\dot{\chi}=\frac{\dot{\phi}_{0}^{2}\sin^{4}\theta_{0}\cos\theta}{\sin^{3}\theta}-\frac{g}{R}\left(\sin\theta\right)\\
\dot{\theta}=\chi\\
\dot{\phi}=\frac{\dot{\phi}_{0}\sin^{2}\theta_{0}}{\sin^{2}\theta}
\end{cases}

First case: initial condition is $\chi_{0}=\dot{\theta}_{0}=\frac{\pi}{10}$, $\theta_{0}=\frac{\pi}{18}$, $\dot{\phi}_{0}=\frac{\pi}{10}$ and $\phi_{0}=0$

```python
from scipy.integrate import odeint
import matplotlib as mpl
import matplotlib.pyplot as plt

from mpl_toolkits.mplot3d import Axes3D
def f_int(y, x, dphi0, thet0):
    g=9.8
    R=1
    return [(dphi0**2)*(np.cos(y[1])*np.sin(thet0)**4)/(np.sin(y[1])**3)+g*np.sin(y[1])/R,y[0],(dphi0*np.sin(thet0)**2)/(np.sin(y[1])**2)]

y_0 = np.array([np.pi/10, np.pi/18, 0.0])

x = np.linspace(0.0, 20.0, 1000000)

y_sol = odeint(f_int, y_0, x, args=(np.pi/10,np.pi/18))

plt.figure(figsize=(9,7),dpi=98)
plt.plot(x, y_sol[:,0],'b')
plt.plot(x, y_sol[:,1],'r')
plt.plot(x, y_sol[:,2],'g')
plt.grid(True)
plt.legend()
plt.show()

fig = plt.figure()
ax = fig.gca(projection='3d')

x1 = np.sin(y_sol[:,1])*np.cos(y_sol[:,2])
y1 = np.sin(y_sol[:,1])*np.sin(y_sol[:,2])
z1 = np.cos(y_sol[:,1])
ax.plot(x1, y1, z1)

us = np.linspace(0, np.pi, 40)
vs = np.linspace(0, 2 * np.pi, 40)
xs = np.outer(np.sin(us), np.sin(vs))
ys = np.outer(np.sin(us), np.cos(vs))
zs = np.outer(np.cos(us), np.ones_like(vs))
ax.plot_surface(xs, ys, zs, linewidth=0)
ax.legend()
fig.tight_layout()
plt.show()
```
