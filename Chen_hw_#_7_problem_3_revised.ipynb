$L=\frac{I_{1}}{2}\left[\dot{\theta}^{2}+\dot{\phi}^{2}\sin^{2}\left(\theta\right)\right]+\frac{I_{3}}{2}\left[\dot{\psi}+\dot{\phi}\cos\left(\theta\right)\right]^{2}-Mgl\cos\theta$

$\frac{\partial L}{\partial\dot{\psi}}=I_{3}\left[\dot{\psi}+\dot{\phi}\cos\left(\theta\right)\right]=I_{3}\omega_{3}=I_{1}a$

$\frac{\partial L}{\partial\dot{\phi}}=\dot{\phi}I_{3}\cos^{2}\theta+\dot{\phi}I_{1}\sin^{2}\left(\theta\right)+I_{3}\dot{\psi}\cos\left(\theta\right)=I_{1}b$

By eliminating $\dot{\psi}$, I got

$\dot{\phi}I_{1}\sin^{2}\left(\theta\right)+I_{1}a\cos\left(\theta\right)=I_{1}b$

$\dot{\phi}=\frac{b-a\cos\left(\theta\right)}{\sin^{2}\left(\theta\right)}$

and 

$\dot{\psi}=\frac{I_{1}}{I_{3}}a-\frac{b-a\cos\left(\theta\right)}{\sin^{2}\left(\theta\right)}\cos\left(\theta\right)$

The Hamtonian is

$H=T+V=\frac{I_{1}\dot{\theta}^{2}}{2}+\frac{I_{1}\left(b-a\cos\theta\right)^{2}}{2\sin^{2}\left(\theta\right)}+\frac{I_{3}\omega_{3}^{2}}{2}+Mgl\cos\theta$

Let

$\alpha=\frac{2H-I_{3}\omega_{3}^{2}}{I_{1}}$

$\beta=\frac{2Mgl}{I_{1}}$

$a=\frac{p_{\psi}}{I_{1}}$

$b=\frac{p_{\phi}}{I_{1}}$

$\frac{\alpha I_{1}}{2}=\frac{I_{1}\dot{\theta}^{2}}{2}+\frac{I_{1}\left(b-a\cos\theta\right)^{2}}{2\sin^{2}\left(\theta\right)}+\frac{\beta I_{1}\cos\theta}{2}$

$\alpha=\dot{\theta}^{2}+\frac{\left(b-a\cos\theta\right)^{2}}{\sin^{2}\left(\theta\right)}+\beta\cos\theta$

Let $u=\cos\theta$. I got 

$\dot{u}^{2}=\left(1-u^{2}\right)\left(\alpha-\beta u\right)-\left(b-au\right)^{2}$

Initially $\alpha=\beta\cos\theta_{0}$ and$ u_{0}=\frac{b}{a}$,

$\dot{u}^{2}=\beta\left(1-u^{2}\right)\left(u-u_{0}\right)-a^{2}\left(u-u_{0}\right)^{2}$

Assume $u_{1}$is another root of $\beta\left(1-u^{2}\right)\left(u-u_{0}\right)-a^{2}\left(u-u_{0}\right)^{2}$,

$\beta\left(1-u_{1}^{2}\right)-a^{2}\left(u_{1}-u_{0}\right)=0$

$u_{1}^{2}+\frac{a^{2}}{\beta}u_{1}-\frac{a^{2}}{\beta}u_{0}-1=0$

Let $u_{1}-u_{0}=x_{1}$. We got 

$x_{1}^{2}+px_{1}+q=0$

which

$p=\frac{a^{2}}{\beta}-2\cos\theta_{0}$

$q=\sin^{2}\theta_{0}$

For top fast problem, $p\gg q$, I got

$x_{1}\approx\frac{q}{p}\approx\frac{I_{1}}{I_{3}}\frac{2Mgl}{I_{3}\omega_{3}^{2}}\sin^{2}\theta_{0}$

Let $u-u_{0}=x$ and $1-u^{2}$as the initial value $\sin^{2}\theta_{0}$ 

$\left(\dot{x}\right)^{2}=a^{2}x\left(x_{1}-x\right)$

Let $y=x-\frac{x_{1}}{2}$

$\left(\dot{y}\right)^{2}=a^{2}\left(\frac{x_{1}^{2}}{4}-y^{2}\right)$

By derivativating the equation again, I can get

$\ddot{y}=-ay$

Initially, x=0. The solution becomes

$x=\frac{x_{1}}{2}\left(1-\cos\frac{I_{3}}{I_{1}}\omega_{3}t\right)$

Final solution is

$\cos\theta-\cos\theta_{0}=\frac{I_{1}}{I_{3}}\frac{Mgl}{I_{3}\omega_{3}^{2}}\sin^{2}\theta_{0}\left(1-\cos\frac{I_{3}}{I_{1}}\omega_{3}t\right)$

$\dot{\phi}=\frac{Mgl}{I_{3}\omega_{3}}\left(1-\cos\frac{I_{3}}{I_{1}}\omega_{3}t\right)$

$\phi=\frac{Mgl}{I_{3}\omega_{3}}\left(t-\frac{I_{1}}{I_{3}\omega_{3}}\sin\frac{I_{3}}{I_{1}}\omega_{3}t\right)+\phi_{0}$

$\dot{\psi}\simeq\omega_{3}+\cos\theta_{0}\frac{Mgl}{I_{3}\omega_{3}}\left(1-\cos\frac{I_{3}}{I_{1}}\omega_{3}t\right)$

$\psi\simeq\omega_{3}t+\cos\theta_{0}\frac{Mgl}{I_{3}\omega_{3}}\left(t-\frac{I_{1}}{I_{3}\omega_{3}}\sin\frac{I_{3}}{I_{1}}\omega_{3}t\right)+\psi_{0}$

```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import symmetric_top as sy
```

$I_{3}\omega_{3}^{2}=I3os$

$Mgl=M$


```python
def theta(M,o3,I1,I3,th0,t):
    I3os=I3*o3*o3
    th=I1*M*(np.sin(th0)**2)*(1-np.cos(I3*o3*t/I1))/(I3os*I3)
    return np.arccos(np.cos(th0)-th)
    
def phi(M,o3,I1,I3,ph0,t):
    I3os=I3*o3*o3
    a=I3*o3/I1
    ph=M*(t-np.sin(a*t)/a)*o3/(I3os)+ph0
    return ph

def psi(M,o3,I1,I3,ps0,th0,t):
    I3os=I3*o3*o3
    a=I3*o3/I1
    ps=o3*t+o3*np.cos(th0)*M*(t-np.sin(a*t)/a)/(I3os)+ps0
    return ps
    
```

```python
M=0.001
o3=1
I1=1
I3=1
th0=np.pi/10
ph0=0
ps0=0
t=np.linspace(0,6*3.1475778061859545,1000)
```

```python
cond = np.array([2*M, th0, 0, o3, I3/I1])
num_pers = 3
beta = cond[0]
theta_0 = cond[1]
d_theta_0 = cond[2]
d_psi_0 = cond[3]
i_rat = cond[4]
alpha = d_theta_0**2+beta*np.cos(theta_0)
a = i_rat*d_psi_0
b = a*np.cos(theta_0)
st = sy.SymTopIniCond(alpha, beta, a, b, i_rat)
```

```python
num_pnts = 1000
num_periods = num_pers
sol = st.solv_eqn(num_pnts, num_periods)
```

```python
fig, ax = plt.subplots(3, sharex=True)
var = [r'$\phi$', r'$\theta$', r'$\psi$']
#for ind in range(3):
ax[0].plot(st.time, sol[:, 0])
ax[0].plot(t,phi(M,o3,I1,I3,ph0,t),'r')
yl = ax[0].set_ylabel(var[0])
yl.set_rotation(0)

ax[1].plot(st.time, sol[:, 1])
ax[1].plot(t,theta(M,o3,I1,I3,th0,t),'g')
yl = ax[1].set_ylabel(var[1])
yl.set_rotation(0)

ax[2].plot(st.time, sol[:, 2])
ax[2].plot(t,psi(M,o3,I1,I3,ps0,th0,t),'b')
yl = ax[2].set_ylabel(var[2])
yl.set_rotation(0)

ax[2].set_xlabel('time')
plt.show()

fig, ax = plt.subplots(3, sharex=True)
ax[0].plot(st.time, sol[:, 0]-phi(M,o3,I1,I3,ph0,t))
yl = ax[0].set_ylabel(var[0])
yl.set_rotation(0)

ax[1].plot(st.time, sol[:, 1]-theta(M,o3,I1,I3,th0,t))
yl = ax[1].set_ylabel(var[1])
yl.set_rotation(0)

ax[2].plot(st.time, sol[:, 2]-psi(M,o3,I1,I3,ps0,th0,t))
yl = ax[2].set_ylabel(var[2])
yl.set_rotation(0)
plt.show()

```
